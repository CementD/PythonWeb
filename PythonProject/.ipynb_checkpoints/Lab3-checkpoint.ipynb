#### У файлі CSV збережено дані про фільми: "Назва", "Жанр", "Рік", "Рейтинг".
#### Прочитайте дані з CSV у DataFrame.
#### Виведіть перші п'ять рядків таблиці.
#### Визначте загальну кількість рядків і стовпців.
#### Визначте типи даних для кожного стовпця.

```python
df_from_file = pd.read_csv("films.csv")
print(df_from_file.head())
```

```python
for col in df_from_file.columns:
    none_values = 0
    for value in df_from_file[col]:
        if value is None:
            none_values += 1
print(f"Count of nan values: {none_values}")
```

# Task5

#### Є список студентів із зазначенням: "Ім'я", "Курс", "Оцінка".
#### Створіть DataFrame і заповніть його не менш ніж вісьмома рядками.
#### Виведіть вихідний DataFrame.
#### Обчисліть середню оцінку за кожним курсом.
#### Знайдіть максимальну та мінімальну оцінку за кожним курсом.

```python
df5 = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Ethan", "Fiona", "George", "Hannah"],
    "Course": [1, 2, 1, 3, 2, 3, 1, 2],
    "Grade": [85, 90, 78, 92, 88, 75, 95, 80]
})
df5
```

```python
course_av = df5.groupby("Course")["Grade"].mean()
course_min = df5.groupby("Course")["Grade"].min()
course_max = df5.groupby("Course")["Grade"].max()
print(f"Course averages: \n{course_av} \nCourse max: \n{course_max} \nCourse min: \n{course_min}")
```

# Task6

#### Зібрано дані про товари та продажі у двох таблицях. Перша містить: "ID", "Назва", "Ціна". Друга містить: "ID", "Кількість".
#### Створіть два DataFrame і заповніть їх не менш ніж п'ятьма рядками.
#### Виведіть обидва DataFrame.
#### Об'єднайте їх за колонкою "ID".
#### Додайте новий стовпець "Сума" як добуток ціни та кількості.
#### Виведіть підсумковий DataFrame.

```python
products_df = pd.DataFrame({
    "ID": [1, 2, 3, 4, 5],
    "Name": ["Laptop", "Phone", "Tablet", "Monitor", "Keyboard"],
    "Price": [1200, 800, 500, 300, 50]
})

sales_df = pd.DataFrame({
    "ID": [1, 2, 3, 4, 5],
    "Quantity": [5, 10, 7, 6, 20]
})
products_df
```

```python
sales_df
```

```python
m_df = pd.merge(products_df, sales_df, on='ID')
m_df
```

```python
m_df["Sum"] = m_df["Price"] * m_df["Quantity"]
m_df
```
