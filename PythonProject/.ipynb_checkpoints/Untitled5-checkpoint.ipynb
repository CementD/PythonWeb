#### Виведіть вихідний масив.
#### Запитайте у користувача число (порогове значення).

# Task2

#### Запитайте в користувача верхню та нижню межу діапазону чисел.
#### Створіть квадратну матрицю 5×5 з цілими значеннями в цьому діапазоні.
#### Виведіть вихідну матрицю.
#### Отримайте головну діагональ і виведіть діагональні елементи.
#### Підрахуйте суму елементів головної діагоналі та виведіть результат.
#### Обнуліть елементи строго вище головної діагоналі та виведіть змінену матрицю.

```python
n1 = int(input("Enter lowest value: "))
n2 = int(input("Enter highest value: "))
arr2 = np.random.randint(n1, n2 + 1, size=(5, 5))
print(arr2)
```

```python
diag = []
for i in range(5):
    diag.append(int(arr2[i][i]))
print(diag)
```

```python
sum = 0
for v in diag:
    sum += v
print(f'Diagonal sum: {sum}')
```

```python
for i in range(5):
    for j in range(5):
        if j > i:
            arr2[i][j] = 0
print(arr2)
```

# Task3

#### Запитайте у користувача початок і кінець діапазону чисел.
#### Сформуйте послідовність цілих чисел у вказаному діапазоні.
#### Перетворіть послідовність на матрицю розміром 6×5 і виведіть її.
#### Підрахуйте суму елементів у кожному рядку і виведіть отриманий одновимірний масив сум.
#### Визначте максимальні значення за кожним стовпчиком і виведіть отриманий одновимірний масив максимумів.

```python
s = int(input("Enter start: "))
e = int(input("Enter end: "))
arr3 = np.arange(s, e+1)
```

```python
print(arr3)
```

```python
if len(arr3) < 30:
    zeros = np.zeros(30 - len(arr3))
    arr3_2 = np.concatenate((arr3, zeros))
elif len(arr3) > 30:
    arr3_2 = arr3[:30]
else:
    arr3_2 = arr3

matrix = arr3_2.reshape(6, 5)
print(matrix)
```

```python
print(f'Sum in rows: {matrix.sum(axis=1)}')
print(f'Max in columns: {matrix.max(axis=0)}')
```

#### Запитайте в користувача довжину масивів.
#### Створіть два одновимірні масиви вказаної довжини:
#### перший заповніть випадковими числами від 0 до 10;
#### другий заповніть випадковими числами від 10 до 20.
#### Виведіть обидва вихідні масиви.
#### Об'єднайте масиви в один і виведіть результат.
#### Виконайте поелементне додавання та поелементну різницю вихідних масивів і виведіть обидва результати.

```python
n = int(input("Enter n: "))
arr5_1 = np.random.randint(0, 11, size=n)
arr5_2 = np.random.randint(11, 21, size=n)
print(f"Arr1: {arr5_1} \nArr2: {arr5_2}")
```

```python
arr5_3 = np.concatenate((arr5_1, arr5_2))
print(f'Merged arr: {arr5_3}')
```

# Task6

#### Запитайте у користувача розмірність вихідної матриці (наприклад, 4×6).
#### Створіть матрицю з послідовно зростаючих цілих чисел заданої розмірності.
#### Виведіть вихідну матрицю.
#### Запитайте в користувача нові розміри (наприклад, 2×12) і перетворіть матрицю в зазначену форму.
#### Виведіть отриману матрицю.
#### Визначте мінімальні та максимальні значення в кожному рядку нової матриці та виведіть обидва одновимірні масиви.
#### Підрахуйте загальну суму всіх елементів нової матриці та виведіть результат.

```python
r = int(input("Enter rows: "))
c = int(input("Input cols: "))
arr6 = np.arange(1, r*c+1).reshape(r, c)
print(arr6)
```

```python
nr = int(input("Enter new rows: "))
nc = int(input("Input new cols: "))
new_arr6 = arr6.reshape(nr, nc)
print(new_arr6)
```

```python
print(f"Mins: {new_arr6.min(axis=1)}")
print(f"Maxs: {new_arr6.max(axis=1)}")
```

```python
print(f"Sum: {new_arr6.sum()}")
```
