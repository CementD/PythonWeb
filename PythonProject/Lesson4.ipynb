```python
month = sales["month"]
revenue = sales["revenue"]

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(month, revenue)
ax.set_title("Revenue annual")
ax.set_xlabel("Month")
ax.set_ylabel("Revenue, (thousand $)")
plt.show()
```

```python
month = sales["month"]
revenue = sales["revenue"]

fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(month, revenue)
ax.set_title("Revenue annual")
ax.set_xlabel("Month")
ax.set_ylabel("Revenue, (thousand $)")
plt.show()
```

```python
month = sales["month"]
revenue = sales["revenue"]

fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(month, revenue,
       color="red",
       linewidth=3,
       linestyle=":",
       marker="o",
       markersize=7,
       markerfacecolor="white",
       markeredgewidth=3 
       )
ax.set_title("Revenue annual", fontsize=16, fontweight="bold")
ax.set_xlabel("Month", fontsize=16)
ax.set_ylabel("Revenue, (thousand $)", fontsize=12)
plt.show()
```

```python
month = sales["month"]
revenue_2023 = sales["revenue"]
revenue_2024 = [140, 154, 135, 160, 180, 170, 140, 190, 210, 200, 220, 205]

fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(month, revenue_2023, marker="o", label="2023") # markers: o, s, ^, d, *
ax.plot(month, revenue_2024, marker="s", label="2024")
ax.set_title("Revenue annual")
ax.set_xlabel("Month")
ax.set_ylabel("Revenue, (thousand $)")
ax.legend()
plt.show()
```

```python
products = ["Phone", "Laptop", "Earphones", "Case"]
sales = [350, 120, 280, 450]
fig, ax = plt.subplots(figsize=(8, 4))
ax.bar(products, sales, color="coral")
ax.set_title("Sales in categories")
ax.set_xlabel("Count")
plt.show()
```

```python
products = ["Phone", "Laptop", "Earphones", "Case"]
sales = [350, 120, 280, 450]
fig, ax = plt.subplots(figsize=(8, 4))
ax.barh(products, sales, color="coral")
ax.set_title("Sales in categories")
ax.set_xlabel("Count")
plt.show()
```

```python
products = ["Phone", "Laptop", "Earphones", "Case"]
sales_2023 = [350, 120, 280, 450]
sales_2024 = [400, 150, 250, 400]

x = np.arange(len(products))
width = 0.35

fig, ax = plt.subplots(figsize=(9, 4))
ax.bar(x - width / 2, sales_2023, width, label="2023")
ax.bar(x + width / 2, sales_2024, width, label="2024")
ax.set_xticks(x)
ax.set_xticklabels(products)
ax.legend()
ax.set_title("Sales comparison in categories 2023 vs 2024")
ax.set_xlabel("Count")
plt.show()
```

```python
products = ["Phone", "Laptop", "Earphones", "Case"]
sales_2023 = [350, 120, 280, 450]
sales_2024 = [400, 150, 250, 400]

x = np.arange(len(products))
width = 0.35

fig, ax = plt.subplots(figsize=(9, 4))
ax.barh(x - width / 2, sales_2023, width, label="2023")
ax.barh(x + width / 2, sales_2024, width, label="2024")
ax.set_xticks(x)
ax.set_xticklabels(products)
ax.legend()
ax.set_title("Sales comparison in categories 2023 vs 2024")
ax.set_xlabel("Count")
plt.show()
```

```python
products = ["Phone", "Laptop", "Earphones", "Case"]
sales = [350, 120, 280, 450]
fig, ax = plt.subplots(figsize=(8, 4))
colors = ["red", "blue", "coral", "green"]

ax.bar(products, sales, color=colors)
ax.set_title("Sales in categories")
ax.set_xlabel("Count")
plt.show()
```

```python
products = ["Phone", "Laptop", "Earphones", "Case"]
sales = [350, 120, 280, 450]
fig, ax = plt.subplots(figsize=(6, 6))
colors = ["red", "blue", "coral", "green"]

ax.pie(sales, 
       labels=products,
       autopct="%1.1f%%",
       startangle=90,
       colors=["red", "blue", "coral", "green"])
ax.set_title("Sales in categories")
ax.set_xlabel("Count")
plt.show()
```
