## Звіт із продажів

### У цьому розділі ми рахуємо підсумки за простими даними

```python
sales = [120, 135, 100, 150, 90, 200, 170]
total = sum(sales)
average = total / len(sales)
maximum = max(sales)
minimum = min(sales)
```

## Оцінки студентів

##### Оцінки студентів та підрахування всіх операцій

```python
students = [("Alex", 85), ("Maria", 92), ("John", 74), ("Olga", 60), ("Max", 48)]
scores = []
for st in students:
    scores.append(st[1])

average = sum(scores) / len(scores)
maximum = max(scores)
minimum = min(scores)
low_scores = [st for st in students if st[1] < 60]

print("Середній бал:", average)
print("Максимум:", maximum)
print("Мінімум:", minimum)
print("Нижче 60:", low_scores)

for name, score in students:
    print(name, "|", "*" * (score // 10))
```

| Середній бал | Максимальний бал | Мінімальний бал |
|--------------|------------------|-----------------|
|    71,8      |        92        |        48       |
