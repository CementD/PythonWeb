```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
company_a = [3000, 3200, 3100, 3300, 3150, 50000]
company_b = [4500, 4800, 4600, 4700, 4550, 4650]

print(np.mean(company_a))
print(np.mean(company_b))
```

```python
print(np.median(company_a))
print(np.median(company_b))
```

```python
data = np.array([10, 20, 30, 40, 50])
manual_mean = sum(data) / len(data)
print(f"Manual (sum/len): {manual_mean}")
print(f"Numpy (mean): {np.mean(data)}")
print(f"Numpy (median): {np.median(data)}")
print(f"Pandas: {pd.Series(data).mean()}")
```

```python
data_odd = np.array([10, 20, 30, 40, 50])
data_even = np.array([10, 20, 30, 40])
print(f"Odd: {np.median(data_odd)}")
print(f"Even: {np.median(data_even)}")
```

```python
ratings = [5, 4, 4, 5, 3, 4, 5, 5, 6, 6, 4, 3, 5, 3]
mode_result = stats.mode(ratings)
print(f"Mode: {mode_result.mode}")
print(f"Mode: {mode_result.count}")
s = pd.Series(ratings)
print(f"Mode: {s.mode().values}")
print(f"Mode: {s.value_counts()}")
```

```python
group_a = np.array([48, 49, 50, 51, 52])
group_b = np.array([20, 35, 50, 65, 80])

print(f"A: mean={group_a.mean()}, std={group_a.std():.2f}")
print(f"B: mean={group_b.mean()}, std={group_b.std():.2f}")
```

```python
np.random.seed(42)
data = np.random.normal(100, 15, 200)
q1 = np.percentile(data, 25)
q2 = np.percentile(data, 50)
q3 = np.percentile(data, 75)

iqr = q3 - q1

print(f"Q1 (25%): {q1:.2f}")
print(f"Q2 (50%): {q2:.2f}")
print(f"Q3 (75%): {q3:.2f}")
print(f"IQR: {iqr:.2f}")
```

```python
for p in [5, 10, 25, 50, 75, 90, 95]:
    val = np.percentile(data, p)
    print(f"{p} : {val:.2f}")
```
