# TASK4

```python
subjects = [("maths", [80, 92, 75, 60, 88]), ("physics", [70, 65, 90, 85, 78]), ("english", [95, 100, 85, 90, 87])]
for name, scores in subjects:
    avg = sum(scores) / len(scores)
    mx = max(scores)
    mn = min(scores)
    
    print(f"{name}: avg={avg}, max={mx}, min={mn}")

best_avg = max(subjects, key=lambda x: sum(x[1]) / len(x[1]))
print(best_avg)
```
