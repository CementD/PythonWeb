```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Task1

###### У вас є список чисел від 1 до 10 та їхні квадрати. Потрібно побудувати графік залежності квадрата числа від самого числа.

###### Створіть два списки: перший містить числа від 1 до 10, другий — їхні квадрати.
###### Побудуйте лінійний графік, де по осі X відкладаються числа, а по осі Y — їхні квадрати.
###### Додайте заголовок "Графік функції y = x²" і підписи осей "X" і "Y".
###### Зробіть лінію червоного кольору та позначте кожну точку маркером.
###### Відобразіть графік на екрані.

```python
arr1 = np.arange(1, 11)
arr2 = arr1 ** 2

fig, ax = plt.subplots()
ax.plot(arr1, arr2,
       color="red",
       marker="o",
       markersize=7,
       markerfacecolor="white"
       )
ax.set_title("Графік функції y = x²")
ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.show()
```

# Task2

###### Є дані про продажі за 6 місяців: [150, 200, 250, 300, 280, 320]. Потрібно показати, як змінювалися продажі за місяцями.

###### Створіть список місяців і список продажів.
###### Побудуйте стовпчасту діаграму: по осі X — місяці, по осі Y — продажі.
###### Додайте підписи осей і заголовок діаграми.
###### Зробіть стовпці синього кольору.
###### Відобразіть діаграму на екрані.

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [150, 200, 250, 300, 280, 320]
fig, ax = plt.subplots(figsize=(8, 4))
ax.bar(months, sales, color="blue")
ax.set_title = "Sales for the half year"
ax.set_xlabel("Month")
ax.set_ylabel("Sales")
plt.show()
```

# Task3

###### Дано три категорії товарів: "Продукти", "Одяг", "Техніка" та їхні частки продажів [40, 30, 30]. Потрібно показати розподіл продажів між категоріями.

###### Створіть список категорій і список їхніх часток.
###### Побудуйте кругову діаграму.
###### Додайте підписи до кожного сектора.
###### Покажіть процентні значення всередині кругової діаграми.
###### Відобразіть результат.

```python
categories = ["Products", "Clothes", "Technic"]
sales = [40, 30, 30]
fig, ax = plt.subplots(figsize=(6, 6))
ax.pie(sales, 
       labels=categories,
       autopct="%1.1f%%",
       startangle=90,
       colors=["red", "blue", "green"])
plt.show()
```

# Task4

###### У вас є дані про зріст і вагу п'яти осіб: зріст [160, 165, 170, 175, 180], вага [55, 60, 70, 72, 80]. Потрібно показати залежність ваги від зросту.

###### Створіть список зростів і список ваг.
###### Побудуйте точковий графік (scatter plot), де по осі X — зріст, а по осі Y — вага.
###### Додайте підписи осей і заголовок "Залежність ваги від зросту".
###### Зробіть точки зеленого кольору.
###### Відобразіть графік на екрані.

```python
growth = [160, 165, 170, 175, 180]
weight = [55, 60, 70, 72, 80]
fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(growth, weight,
       color="green",
       linewidth=3,
       linestyle=":",
       marker="o",
       markersize=7,
       markerfacecolor="green",
       markeredgewidth=3 
       )
ax.set_title("Залежність ваги від зросту")
ax.set_xlabel("Growth")
ax.set_ylabel("Weight")
plt.show()
```

# Task5

###### Є дані про температуру протягом доби: години [0, 4, 8, 12, 16, 20, 24], температура [15, 14, 18, 22, 25, 20, 17]. Потрібно побудувати графік зміни температури за годинами.

###### Створіть список годин і список температур.
###### Побудуйте лінійний графік залежності температури від часу.
###### Додайте заголовок "Температура протягом доби" і підписи осей.
###### Увімкніть відображення сітки на графіку.
###### Відобразіть результат.

```python
hours = [0, 4, 8, 12, 16, 20, 24]
temps = [15, 14, 18, 22, 25, 20, 17]
fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(hours, temps,
       color="red",
       marker="o",
       markersize=7,
       markerfacecolor="white"
       )
ax.set_title("Температура протягом доби")
ax.set_xlabel("Hour")
ax.set_ylabel("Temp")
plt.show()
```

# Task6

###### Є таблиця з результатами тестів п'яти студентів: "Alex" – 85, "Irene" – 92, "John" – 78, "Mary" – 90, "Boris" – 88. Потрібно показати результати студентів на діаграмі.

###### Створіть список імен і список результатів.
###### Побудуйте стовпчасту діаграму: по осі X — імена студентів, по осі Y — їхні результати.
###### Додайте заголовок "Результати тесту" та підписи осей.
###### Зробіть стовпці помаранчевого кольору.
###### Відобразіть діаграму на екрані.

```python
results = pd.DataFrame({
    "name": ["Alex", "Irene", "John", "Mary", "Boris"],
    "grade": [85, 92, 78, 90, 88]
})
fig, ax = plt.subplots()
ax.bar(results["name"], results["grade"], color="orange")
ax.set_title("Test results")
ax.set_xlabel("Student")
ax.set_ylabel("Grade")
plt.show()
```
